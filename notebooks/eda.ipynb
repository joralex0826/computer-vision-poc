```python
infractions = data[data["infraction_detected"]].reset_index(drop=True)

print(len(infractions))
infractions.head(20)
```

```python
from IPython.display import HTML

def review_infractions(df, n=20, cols=("picture_url", "labels_detected", "ocr_text")):
    sample = df.head(n).copy()
    sample["imagen"] = sample["picture_url"].apply(lambda u: f'<img src="{u}" width="120">')
    sample["picture_url"] = sample["picture_url"].apply(lambda u: f'<a href="{u}" target="_blank">{u}</a>')
    return HTML(sample[["imagen", *cols]].to_html(escape=False))

review_infractions(infractions, n=20)
```
